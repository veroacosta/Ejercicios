# Cuadratica

En la ecuación $x^2+px+q=0$ y $q$ se toman aleatoriamente en $[0,2]$. ¿Cúal es la probabilidad de que sus raíces sean reales?

# 1. Identificando las variables
Tenemos dos: $p$ y $q$

# 2 y 3. Determinando la distribución de probabilidad y modelando las variables aleatorias
Se especifica que $p $~$ U(0,2)\ y\ que\ $q $~$ U(0,2)$ 

# 4. Definiendo el modelo y los objetivos
El determinante $p^2-4*a*q$ tendra que ser positivo para que el resultado obtenido sea una raíz real, es decir $x_2-x_1\leq{0}$



```python
import numpy as np
import random
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
def discriminante():
  d=0
  p=np.random.uniform(0,2)
  q=np.random.uniform(0,2)
  if p ** 2 - 4 * q >= 0:
    d += 1
  return d
```

```python
discriminante()
```

```python
def prob_d(n):
  prob = []
  d = 0
  for i in range(n):
    d += discriminante()
    prob.append(d/(i+1))
  return prob
```

```python
X = prob_d(1000)
```

```python
plt.plot(X)
```

```python
def modelo(n):
  lis = []
  for i in range(n):
    s = prob_d(1000)
    lis.append(np.mean(s))
  return lis
```

```python
Y = modelo(1000)
```

```python
plt.hist(Y)
plt.show()
```

```python
print(np.mean(Y), np.std(Y))
```
